```python
def detect_and_plot_outliers(data):
    for col in data.select_dtypes(include='number').columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # Detect outliers
        outliers = data[(data[col] < lower) | (data[col] > upper)][col]

        print(f"{col}: {len(outliers)} outliers")

        # Only plot if outliers are present
        if not outliers.empty:
            plt.figure(figsize=(6, 4))
            sns.boxplot(y=outliers)
            plt.title(f'Outliers in {col}')
            plt.ylabel(col)
            plt.grid(True)
            plt.show()
```

```python
detect_and_plot_outliers(data)
```

```python
def remove_outliers_iqr(data):
    cleaned_data = data.copy()
    for col in cleaned_data.select_dtypes(include='number').columns:
        Q1 = cleaned_data[col].quantile(0.25)
        Q3 = cleaned_data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # Keep only the non-outlier values
        cleaned_data = cleaned_data[(cleaned_data[col] >= lower) & (cleaned_data[col] <= upper)]

    return cleaned_data
```

```python
cleaned_data = remove_outliers_iqr(data)
```

```python
for col in cleaned_data.select_dtypes(include='number').columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=cleaned_data[col])
    plt.title(f'Boxplot for {col} (Outliers Removed)')
    plt.ylabel(col)
    plt.grid(True)
    plt.show()
```

```python
categorical_cols = ['Fulfilment', 'ship-service-level', 'Category', 'Courier Status']

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x=col, order=data[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
shipped_status_cols = [
     'Status_Shipped - Damaged',
    'Status_Shipped - Delivered to Buyer', 'Status_Shipped - Lost in Transit',
    'Status_Shipped - Out for Delivery', 'Status_Shipped - Picked Up',
    'Status_Shipped - Rejected by Buyer', 'Status_Shipped - Returned to Seller',
    'Status_Shipped - Returning to Seller'
]

status_counts = data[shipped_status_cols].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
status_counts.plot(kind='bar', color='skyblue')
plt.title("Frequency of Shipped Statuses")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
